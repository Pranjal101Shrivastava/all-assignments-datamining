```python
# --- Install ---
!pip -q install -U autogluon.tabular==1.4.0

# --- Load ---
import pandas as pd
df = pd.read_csv("/kaggle/input/california-housing-prices/housing.csv")
df = df.rename(columns={"median_house_value":"label"})  # set target
print(df.shape); df.head()

# --- Train (fast) ---
from autogluon.tabular import TabularPredictor
pred = TabularPredictor(label="label", problem_type="regression", path="ag_california")
pred.fit(
    train_data=df,
    time_limit=300,                 # 5 min
    presets="medium_quality",
    hyperparameters={"GBM":[{}],"CAT":[{}],"XGB":[{}],"NN_TORCH":[]},
    verbosity=2
)
pred.leaderboard(silent=True).head(10)

# --- Inference demo & save ---
test_pred = pred.predict(df.drop(columns=["label"]).iloc[:10])
test_pred.to_frame("pred").to_csv("/kaggle/working/california_preds_sample.csv", index=False)
print("Saved /kaggle/working/california_preds_sample.csv")
```
